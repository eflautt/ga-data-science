```python
data = pd.read_csv('./train.csv', skipinitialspace=True)

data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

data['Year'] = data.index.year
data['Month'] = data.index.month


store1_data = data[data.Store == 1]

data.head()
```
